### Analyze Data

```python
#load the data
df = pandas.read_csv(exportfile)
df['created_time'] = pandas.to_datetime(df.created_at,format="%a %b %d %H:%M:%S +0000 %Y")\
                            .apply(lambda moment: moment.tz_localize('UTC').tz_convert("US/Eastern"))
df['user_created_time'] = pandas.to_datetime(df.user_created_at,format="%a %b %d %H:%M:%S +0000 %Y")\
                            .apply(lambda moment: moment.tz_localize('UTC').tz_convert("US/Eastern"))
df['counter'] = 1
print "Rows: " + str(df.shape[0]) + " Columns: " + str(df.shape[1])
```

```python
#deanon the relevant columns and export
key = pandas.read_csv(decoderfile)
key['user_anon'] = key['user_anon'].apply(lambda row: row)
key = key.set_index('user_anon')['user'].to_dict()
df['from_user'] = df['from_user'].apply(lambda user: hp.replacer(user,key))
df['connection'] = df['connection'].apply(lambda user: hp.replacer(user,key))
df['text'] = df['text'].apply(lambda user: hp.text_replacer(user,key))
df['in_reply_to_screen_name'] = df['in_reply_to_screen_name'].apply(lambda user: hp.replacer(user,key))
df.to_csv("data_raw.csv")
```

```python
#split the data into 3 dfs and expor the graph for R
tweets = df[df['type'] == "tweet"]
tweets = tweets.drop_duplicates('id')
hashtag = df[df['type'] == "tag"]
person = df[df['type'] == "user"]
network = df[df['type'] != "tag"]
del df
print "This database contains " + str(tweets.shape[0]) + " tweets"
```

```python
#create a undirected graph
G=nx.Graph()
G.add_nodes_from(person.drop_duplicates(["from_user"]).apply(lambda row : hp.pandas_node_attributes(row), axis = 1))
G.add_edges_from(list(person.apply(lambda row : hp.pandas_to_edgelist_uni(row), axis = 1)))
```

```python
#create a undirected graph
MDG = nx.MultiDiGraph()
MDG.add_nodes_from(person.drop_duplicates(["from_user"]).apply(lambda row : hp.pandas_node_attributes(row), axis = 1))
MDG.add_edges_from(list(person.apply(lambda row : hp.pandas_to_edgelist(row), axis = 1)))
```

```python
#directed mulitgraph for 
exportgraph = nx.MultiDiGraph()
exportgraph.add_nodes_from(person.drop_duplicates(["from_user"]).apply(lambda row : hp.pandas_node_attributes(row), axis = 1))
exportgraph.add_edges_from(list(person.apply(lambda row : hp.pandas_to_edgelist_uni(row), axis = 1)))
```

```python
#make partitions and export graph
partition = community.best_partition(G)

for n in MDG.nodes_iter():
    exportgraph.node[n]['community'] = partition[n]
    MDG.node[n]['community'] = partition[n]

#export the export graph and del
nx.write_graphml(exportgraph,"export_graph.graphml")
del G,exportgraph
```

```python
#get the partitions and size
partition = pandas.DataFrame(pandas.DataFrame(partition.items(), columns=['user', 'community'])\
            .groupby(['community']).apply(lambda x: list(x.user)), columns = ['users'])
partition['size'] = partition.users.apply(lambda x: len(x))
#order by size and get the top 15
partition = partition.sort(['size'],ascending = False)[:15]
```

```python
for community in range(0,numberofcommunities+1):
    sub_mdg = MDG.subgraph(partition.users.iloc[community])
    hashtags = []
    links = []
    language = []
    text = []
    userlang = []
    followers = []
    timezone = []
    username = []
    tweetid = []
    for edge in sub_mdg.edges_iter():
        for tweet in sub_mdg.get_edge_data(edge[0],edge[1])[0]['tweets']:
            hashtags += [tweet['hashtags']]
            links += [tweet['url']]
            text += [tweet['text']]
            language += [tweet['lang']]
            tweetid += [tweet['id']]
    
    for nodes in [nodes for nodes in sub_mdg.nodes(data=True)]:
        if nodes[1].has_key('user_lang'):
            userlang += [nodes[1]['user_lang']]
        if nodes[1].has_key('followers'):
            followers += [nodes[1]['followers']]
        if nodes[1].has_key('time_zone'):
            timezone += [nodes[1]['time_zone']]
        username += [nodes[0]]
    
    hashtags = pandas.DataFrame(pandas.Series(hashtags).value_counts(),columns=["hashtags"])
    language = pandas.DataFrame(pandas.Series(language).value_counts(),columns=["language"])
    text = pandas.DataFrame(pandas.Series(text).value_counts(),columns=["text"])
    userlang = pandas.DataFrame(pandas.Series(userlang).value_counts(),columns=["userlang"])
    followers = pandas.DataFrame(pandas.Series(followers).value_counts(),columns=["followers"])
    timezone = pandas.DataFrame(pandas.Series(timezone).value_counts(),columns=["timezone"])
    username = pandas.DataFrame(pandas.Series(username).value_counts(),columns=["username"])
    links = pandas.DataFrame(pandas.Series(links).value_counts(),columns=["links"])
    tweetid = pandas.DataFrame(pandas.Series(tweetid).value_counts(),columns=["tweetid"])
    hp.save_xls([hashtags,language,text,userlang,followers,timezone,username,links,tweetid],\
             directory + str(partition.iloc[community].name) + "_community.xlsx")
```

```python
#tag and word freqency
wordfreq = pandas.DataFrame(pandas.Series(regex.findall("\w+"," ".join(tweets.text.tolist()).decode('utf-8'))).value_counts())
tagfreq = pandas.DataFrame(pandas.Series(regex.findall("#\w+"," ".join(tweets.text.tolist()).decode('utf-8'))).value_counts())
wordfreq['len'] = wordfreq.index
wordfreq['len'] = wordfreq['len'].apply(lambda x:len(x))
wordfreq = wordfreq[(wordfreq['len'] > 3) & (wordfreq[0] > 10)]
tagfreq = tagfreq[(tagfreq > 0)]
#person etc freqencies
personfreq = pandas.Series(regex.findall("@\w+"," ".join(tweets.text.tolist()).lower())).value_counts()
urlfreq = pandas.Series(regex.split(" "," ".join(tweets.url.dropna().tolist()).lower())).value_counts()
userlangfreq = tweets.drop_duplicates(['from_user']).user_lang.value_counts()
usertimezone = tweets.drop_duplicates(['from_user']).time_zone.value_counts()
repliedtouser = tweets.in_reply_to_screen_name.value_counts()
repliedtostatus = tweets.in_reply_to_status_id_str.value_counts()
mediaurlfreq = tweets.media_url.value_counts()
langfreq = tweets.lang.value_counts()
mediatypefreq = tweets.user_utc_offset.value_counts()
mediatypefreq = tweets.media_type.value_counts()
sourcefreq = tweets.source.value_counts()
blockedfreq = tweets.withheld_in_countries.value_counts()
```

```python
#set freqency unique connections
freqency = "D"
#get unique people
person = person.set_index(['created_time'],drop=False)
unique_connections = (person.from_user + " and " + person.connection).resample(freqency,how=hp.count_uniques)
#get unique tweeters
tweets = tweets.set_index(['created_time'],drop=False)
unique_tweeters = tweets['from_user'].resample(freqency,how=hp.count_uniques)
#get potentials
network = network.set_index(['created_time'],drop=False)
potential_connections = network[['from_user','connection']].resample(freqency,how=hp.potentials).from_user
#get number of tweets
number_of_tweets = tweets['counter'].resample(freqency,how='sum')
#calculate network stats
daystats = pandas.DataFrame(dict(unique_tweeters = unique_tweeters,
                                 number_of_tweets = number_of_tweets,
                                 unique_connections = unique_connections,
                                 potential_connections=potential_connections))
daystats['Network_Density'] = daystats.unique_connections/(daystats.potential_connections*1.0)
daystats['Network_Density_2'] = daystats.unique_connections/(daystats.unique_tweeters*1.0)
```

```python
#calculate participation or tweeters ..
tweeters = tweets[['from_user','counter']].groupby(['from_user'],as_index=False,sort=True).sum()
tweeters = tweeters.sort('counter',ascending=False)
total_tweeters = tweeters.shape[0] *1.0
morethan1 = tweeters.counter[tweeters.counter > 1].shape[0]/total_tweeters
morethan3 = tweeters.counter[tweeters.counter > 3].shape[0]/total_tweeters
morethan5 = tweeters.counter[tweeters.counter > 5].shape[0]/total_tweeters
```

```python
tweeters['DOS'] = tweeters.from_user.apply(lambda user: hp.is_in_user_bank(user,key.values()))
total_dos = len(set(tweeters[(tweeters.from_user.str.contains("@"))==False].from_user))
total_tweeters = len(set(tweeters.from_user))
```

```python
#calculate best times to tweet
tweets['created_moment'] = tweets.created_time.apply(lambda fulldate: str(fulldate.time()))
tweets['created_moment'] = pandas.to_datetime(tweets.created_moment,format="%H:%M:%S")
tweets.index = tweets['created_moment']
besttime = tweets['counter'].resample("30 Min",how='sum')
besttime=besttime.sort_index()
```
